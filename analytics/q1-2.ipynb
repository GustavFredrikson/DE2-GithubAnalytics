```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../top_languages.csv")
df.head()
```

```python
df = df[df["language"] != "No Language"]
```

```python
sns.set_style("white")
sns.set_context("poster")

# calculate total sum
total = df['count'].sum()

# calculate percentage
df['percentage'] = 100 * df['count'] / total

plt.figure(figsize=(10, 10))
barplot = sns.barplot(x="count", y="language", data=df.head(10))

plt.title("Top 10 Languages by Repository Count")
plt.xlabel("Count")
plt.ylabel("Language")

for i, (value, percentage) in enumerate(zip(df['count'].head(10), df['percentage'].head(10))):
        barplot.text(value, i, ' {:.1f}%'.format(percentage), color='black', va="center")

sns.despine(left=True, bottom=True)
plt.grid(False)

plt.show()
```

```python
df = pd.read_csv("../most_commits.csv")
df.head()
```

```python
top10_most_updated = df.sort_values('commits', ascending=False).head(10)

sns.set_style("white")
sns.set_context("poster")

plt.figure(figsize=(10, 10))
barplot = sns.barplot(x="commits", y="name", data=top10_most_updated)

plt.title("Top 10 Most Updated GitHub Projects")
plt.xlabel("Commits")
plt.ylabel("Project Name")

# Adding commit counts on the bars
for i, value in enumerate(top10_most_updated['commits']):
    barplot.text(value, i, ' {:.0f}'.format(value), color='black', va="center")

sns.despine(left=True, bottom=True)
plt.grid(False)

plt.show()
```

```python
top10_most_updated = df.sort_values('commit_frequency', ascending=False).head(10)

sns.set_style("white")
sns.set_context("poster")

plt.figure(figsize=(10, 10))
barplot = sns.barplot(x="commit_frequency", y="name", data=top10_most_updated)

plt.title("Top 10 Most Frequently Updated GitHub Projects")
plt.xlabel("Commits per Day")
plt.ylabel("Project Name")

# Adding commit counts on the bars
for i, value in enumerate(top10_most_updated['commits']):
    barplot.text(value, i, ' {:.0f}'.format(value), color='black', va="center")

sns.despine(left=True, bottom=True) 
plt.grid(False)

plt.show()
```

```python
top10_most_updated = df.sort_values('commit_frequency', ascending=False).head(10)
top10_most_updated
```

```python
top_commit_frequency = df.sort_values('commit_frequency', ascending=False).head(10)

plt.figure(figsize=(10, 10))
barplot = sns.barplot(x="commit_frequency", y="name", data=top_commit_frequency)
plt.title("Top 10 Projects by Commit Frequency")
plt.xlabel("Commit Frequency")
plt.ylabel("Project Name")

# Adding commit frequencies on the bars
for i, value in enumerate(top_commit_frequency['commit_frequency']):
    barplot.text(value, i, ' {:.2f}'.format(value), color='black', va="center")

sns.despine(left=True, bottom=True)
plt.grid(False)
plt.show()
```
